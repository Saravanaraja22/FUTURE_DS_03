```python
# =========================================
# Task 3: Marketing Funnel & Conversion Analysis
# Google Colab - Single Cell Code
# =========================================

import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------
# Create Dummy Funnel Dataset
# ------------------------------
data = {
    "channel": ["Google Ads", "Facebook Ads", "Email", "Organic"],
    "impressions": [50000, 45000, 30000, 40000],
    "leads": [5000, 4200, 6000, 8000],
    "trials": [2000, 1800, 3500, 4200],
    "conversions": [800, 600, 1500, 2200]
}

df = pd.DataFrame(data)

print("DATASET PREVIEW")
display(df)

# ------------------------------
# Conversion Rates
# ------------------------------
df["lead_rate"] = (df["leads"] / df["impressions"]) * 100
df["trial_rate"] = (df["trials"] / df["leads"]) * 100
df["conversion_rate"] = (df["conversions"] / df["trials"]) * 100

print("\nCONVERSION RATES (%)")
display(df[["channel", "lead_rate", "trial_rate", "conversion_rate"]])

# ------------------------------
# Leads by Channel
# ------------------------------
df.set_index("channel")["leads"].plot(kind="bar", title="Leads by Marketing Channel")
plt.xlabel("Channel")
plt.ylabel("Leads")
plt.show()

# ------------------------------
# Conversions by Channel
# ------------------------------
df.set_index("channel")["conversions"].plot(kind="bar", title="Conversions by Channel")
plt.xlabel("Channel")
plt.ylabel("Conversions")
plt.show()

# ------------------------------
# Funnel Drop-off Visualization
# ------------------------------
funnel_totals = df[["impressions", "leads", "trials", "conversions"]].sum()

funnel_totals.plot(kind="bar", title="Overall Marketing Funnel Performance")
plt.xlabel("Funnel Stage")
plt.ylabel("Count")
plt.show()

# ------------------------------
# Key Insights
# ------------------------------
print("KEY INSIGHTS:")
print("1. Organic channel generates the highest conversions.")
print("2. Email marketing shows strong lead-to-trial conversion.")
print("3. Paid ads experience drop-offs at the trial stage.")
print("4. Funnel optimization is required between trials and conversions.")

# ------------------------------
# Save Dataset for GitHub
# ------------------------------
df.to_csv("marketing_funnel_data.csv", index=False)
print("marketing_funnel_data.csv file created successfully.")
```
